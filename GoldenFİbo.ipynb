```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
data = yf.download("BTC-USD", start="2024-11-10", end="2025-04-09")
```

```python
high = data['High'].max()
low = data['Low'].min()
diff = high - low
```

Güncel grafige baktığımızda verilerin doğru olduğunu görüyoruz.




Ekran görüntüsü 2025-04-09 215326.png

```python
high = float(data['High'].max())
low = float(data['Low'].min())
diff = high - low

levels = {
    "0.236": high - 0.236 * diff,
    "0.382": high - 0.382 * diff,
    "0.5": high - 0.5 * diff,
    "0.618": high - 0.618 * diff,
    "0.786": high - 0.786 * diff
}
```

```python
plt.figure(figsize=(14, 6))
plt.plot(data['Close'], label="BTC-USD Fiyat", color='black')
for level, price in levels.items():
    if level == "0.618":
        plt.axhline(price, label=f"Altın Oran {level}", linestyle="--", color='green')
    elif level == "0.382":
        plt.axhline(price, label=f"Fibonacci {level}", linestyle="--", color='red')
    else:
        plt.axhline(price, label=f"Fibonacci {level}", linestyle="--", color='gray')

plt.title("Altın Oran ve Fibonacci Seviyeleri ile Bitcoin Teknik Analizi")
plt.xlabel("Tarih")
plt.ylabel("Fiyat (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
